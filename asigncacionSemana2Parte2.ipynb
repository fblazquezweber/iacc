```python
inventario = [
    ["Desastre", "Luis Brito", "La Ceiba", "Misterio", 250, 10, "Intrigante"],
    ["La Casa", "Jose Perez", "Los Alpes", "Romance", 100, 5, "Dramatico"],
    ["El Detective", "Pirlo Camonero", "Editorial Risueña", "Comedia", 150, 6, "Muchas risas"]
]
opciones = ["Consultar", "Agregar", "Modificar", "Eliminar", "Listar", "Salir"]
modificaciones = ["Titulo", "Autor", "Editorial", "Genero", "Precio", "Unidades", "Descripcion", "Salir"]

def mostrarMenu():
    print('------Menu opciones------')
    for i in range(len(opciones)):
        print(f'{i+1}. {opciones[i]}')
    print('-------------------------')

def mostrarModificar():
    print('------Menu opciones------')
    for i in range(len(modificaciones)):
        print(f'{i+1}. {modificaciones[i]}')
    print('-------------------------')

def buscarLibro():
    buscar = input('Ingrese el titulo del libro a buscar: ')
    for libro in inventario:
        if libro[0] == buscar:
            print(f'Titulo: {libro[0]}')
            print(f'Autor: {libro[1]}')
            print(f'Editorial: {libro[2]}')
            print(f'Genero: {libro[3]}')
            print(f'Precio: {libro[4]}')
            print(f'Unidades: {libro[5]}')
            print(f'Descripcion: {libro[6]}')
            print(f'El libro se encuentra en el estante número {inventario.index(libro)}')
            return
    print(f'El libro {buscar} se encuentra agotado')

def agregarLibro():
    nuevo_libro = []
    nuevo_libro.append(input('Ingrese el titulo del libro a insertar: '))
    nuevo_libro.append(input('Ingrese el autor del libro a insertar: '))
    nuevo_libro.append(input('Ingrese la editorial del libro a insertar: '))
    nuevo_libro.append(input('Ingrese el género del libro a insertar: '))
    nuevo_libro.append(int(input('Ingrese el precio del libro a insertar: ')))
    nuevo_libro.append(int(input('Ingrese la cantidad disponible para este título: ')))
    nuevo_libro.append(input('Ingrese la descripción del libro a insertar: '))
    inventario.append(nuevo_libro)
    print('Libro agregado exitosamente')

def modificarLibro():
    buscar = input('Ingrese el titulo del libro a modificar: ')
    for libro in inventario:
        if libro[0] == buscar:
            while True:
                mostrarModificar()
                opcion = int(input('Ingrese una opcion: '))
                if opcion == 1:
                    libro[0] = input('Ingrese el nuevo título: ')
                elif opcion == 2:
                    libro[1] = input('Ingrese el nuevo autor: ')
                elif opcion == 3:
                    libro[2] = input('Ingrese la nueva editorial: ')
                elif opcion == 4:
                    libro[3] = input('Ingrese el nuevo género: ')
                elif opcion == 5:
                    libro[4] = int(input('Ingrese el nuevo precio: '))
                elif opcion == 6:
                    libro[5] = int(input('Ingrese la nueva cantidad: '))
                elif opcion == 7:
                    libro[6] = input('Ingrese la nueva descripción: ')
                elif opcion == 8:
                    break
                else:
                    print('Opción no válida')
            return
    print(f'El libro {buscar} se encuentra agotado')

def eliminarLibro():
    eliminar = input('Ingrese el titulo del libro a eliminar: ')
    for libro in inventario:
        if libro[0] == eliminar:
            inventario.remove(libro)
            print(f'El libro {eliminar} ha sido eliminado')
            return
    print(f'El libro {eliminar} se encuentra agotado')

def listarLibros():
    for libro in inventario:
        print(f'Titulo: {libro[0]}')

while True:
    mostrarMenu()
    opcion = int(input('Ingrese una opcion: '))
    if opcion == 1:
        buscarLibro()
    elif opcion == 2:
        agregarLibro()
    elif opcion == 3:
        modificarLibro()
    elif opcion == 4:
        eliminarLibro()
    elif opcion == 5:
        listarLibros()
    elif opcion == 6:
        break
    else:
        print('Opción no válida')

```
